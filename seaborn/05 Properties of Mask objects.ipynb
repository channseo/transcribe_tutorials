## Coordinate properties
### x, y, xmin, xmax, ymin, ymax
좌표 속성은 플롯에서 마크가 그려지는 위치를 결정한다. 일반적으로 x 좌표는 가로 위치이고 y 좌표는 세로 위치이다. 일부 마크는 하나 또는 두 변수에 대해 스팬(예: 최소, 최대) 매개변수화를 허용한다. 다른 마크는 x와 y를 허용하지만 기준선 매개변수를 사용하여 span을 표시할 수도 있다. 레이어의 방향 매개변수에 따라 작동 방식이 결정된다.

변수에 숫자 데이터가 포함되지 않은 경우 배율은 데이터를 화면에 그릴 수 있도록 변환을 적용한다. 예를 들어, Nominal scales는 각 고유 범주에 정수 인덱스를 할당하고 시간 척도는 날짜를 참조 'epoch'으로 나타낸다.



## Color properties
### color, fillcolor, edgecolor
* 모든 마크에 색상을 지정할 수 있으며, 많은 마크가 마크의 '가장자리'와 '채우기'의 색상을 구분한다. 
* 종종 단순히 색상을 사용하면 두 가지를 모두 설정할 수 있지만, 보다 구체적인 속성을 사용하면 더 많은 제어가 가능하다.

* 색상 속성이 매핑된 경우 기본 팔레트는 눈금 유형에 따라 달라진다. Nominal scale은 불연속적이고 정렬되지 않은 색상을 사용하는 반면, continuous scales(시간 척도 포함)는 순차적인 그라데이션을 사용한다.

!! 기본 연속 눈금은 식별성을 개선하기 위해 향후 release에서 변경될 수 있다.
<br/>

색상 눈금은 'viridis', 'rocket' 또는 'deep'과 같이 팔레트 이름으로 매개변수화된다. 일부 색상표 이름에는 단순 그라디언트(예: 'dark:blue') 또는 cubehelix 시스템(예: 'ch:start=.2,rot=-.4') 등의 매개변수가 포함될 수 있다.<br/>
적절한 선택에 대한 지침은 색상표 튜토리얼을 참조해라.
<br/><br/>
연속 눈금은 눈금이 삽입된 색상 튜플로 매개변수화할 수도 있다. 명목 눈금을 사용할 때는 팔레트 이름(필요한 경우 개별적으로 샘플링됨), 개별 색상 값 목록 또는 데이터 값을 색상에 직접 매핑하는 사전을 제공할 수 있다.
<br/><br/>
개별 색상은 다양한 형식으로 지정할 수 있다. 여기에는 현재 색상 주기에 대한 색인 참조('C0'), 한 글자 shorthand('b'), 회색조 값('.4'), RGB 헥스 코드('#4c72b0'), X11 색상 이름('seagreen'), XKCD 색상 조사 이름('pupleish') 등이 포함된다.




### alpha, fillalpha, edgealpha
alpha 속성은 마크의 불투명도를 결정한다. alpha를 낮추면 overplotting 시 밀도를 표현하는 데 도움이 될 수 있다.

alpha 속성을 매핑하면 마크가 겹치지 않는 경우에도 중요도를 전달하고 색상 눈금과 결합하여 두 변수를 나타낼 수 있으므로 유용할 수 있다. <br/>
또한 alpha가 낮은 색상은 채도가 낮게 표시되므로 막대와 같이 채워진 큰 마크의 모양을 개선할 수 있다.
색상과 마찬가지로 일부 마크는 추가 제어를 위해 별도의 edgealpha 및 fillalpha 속성을 정의한다.

## Style properties
### fill
fill 속성은 가장자리와 내부가 구분된 마크와 관련이 있으며 내부가 표시되는지 여부를 결정한다. fill은 boolean상태이다. 채우기는 참 또는 거짓으로만 설정할 수 있다.

<br/><br/>
### marker
marker 속성은 dot mark와 일부 선 mark와 관련이 있습니다. 마커를 지정하는 API는 매우 유연하며, matplotlib API 문서(matplotlib.markers)에 자세히 설명되어 있다.
<br/>
marker는 여러 가지 간단한 문자열 코드를 사용하여 지정할 수 있다.

또한 (num_sides, fill_style, angle) 튜플을 사용하여 프로그래밍 방식으로 생성할 수도 있다.

수학 텍스트 문자 코드('$...$') 및 정점 배열을 포함한 추가 형식은 matplotlib 문서를 참조하기.
<br/><br/>
마커 속성은 nominal scale로 매핑되며, 다른 모양에 고유한 순서는 없다. 축척이 제공되지 않으면 플롯은 프로그래밍 방식으로 적절하게 큰 고유 마커 세트를 생성한다.

이렇게 하면 도형이 구분되지만, 대부분의 경우 한 플롯에 몇 개 이상의 마커가 사용되는 경우 마커를 구분하기 어렵다는 점을 염두에 두어야 한다.
<br/><br/>
### linestyle, edgestyle
선 스타일 속성은 선 마크와 관련이 있으며, 가장자리 스타일 속성은 'edgestyle'가 있는 여러 마크와 관련이 있다. 두 속성 모두 온오프 세그먼트 측면에서 선의 'dashing'를 결정합니다.

대시는 소수의 속기 코드('-', '--', '-.', ':')로 지정하거나 프로그래밍 방식으로 (on, off, ...) 튜플을 사용하여 지정할 수 있다. 튜플 사양에서 단위는 선폭과 같다.


## Size properties
### pointsize
point size 속성은 개별 데이터 포인트에 마커를 표시할 수 있는 도트 마크 및 선 마크와 관련이 있다. 단위는 포인트 단위로 마크의 지름에 해당한다.
<br/>
기본적으로 포인트 크기는 데이터의 제곱근에 따라 배율이 조정되므로 크기가 면적이 아닌 직경으로 표시된다.


### linewidth
linewidth 속성은 선 마크와 관련이 있으며 선 마크의 두께를 결정한다. 값은 음수가 아니어야 하며 포인트 단위여야 한다.


### edgewidth
가장자리 너비 속성은 선 너비와 비슷하지만 선이 아닌 가장자리/채우기가 있는 마크에 적용된다. 또한 눈금에 사용할 때 기본 범위가 다릅니다. 단위는 동일하다.


### stroke
stroke 속성은 가장자리 폭과 비슷하지만 도트 마크가 채우기가 아닌 획으로 정의될 때 적용된다. 또한 기본 스케일 범위가 약간 다르지만 그 외에는 비슷하게 작동한다.


## Text properties
### halign, valign
정렬 및 맞춤 속성은 텍스트 마크의 가로 및 세로 정렬을 제어g한다. 가로 맞춤 옵션은 '왼쪽', '오른쪽', '가운데'이며 세로 맞춤 옵션은 '위쪽', '아래쪽', '가운데', '기준선', 'center_baseline'이다.

### fontsize
글꼴 크기 속성은 텍스트 마크의 크기를 제어한다. 값에는 포인트 단위가 있다.

### offset
오프셋 속성은 텍스트 마크와 앵커 위치 사이의 간격을 제어한다. 중앙 맞춤을 사용하지 않을 때(즉, 왼쪽/오른쪽 또는 위쪽/아래쪽을 사용할 때) 적용된다. 값에는 포인트 단위가 있다.


## Text properties
### halign, valign
정렬 및 맞춤 속성은 텍스트 마크의 가로 및 세로 정렬을 제어한다. 가로 맞춤 옵션은 'left', 'right', 'center'이고 세로 맞춤 옵션은 'top', 'bottom', 'center', 'baseline', and 'center_baseline'이다.

### fontsize
글꼴 크기 속성은 텍스트 마크의 크기를 제어한다. 값에는 포인트 단위가 있다.

### offset
offset 속성은 텍스트 마크와 앵커 위치 사이의 간격을 제어한다. 중앙 맞춤을 사용하지 않을 때(즉, 왼쪽/오른쪽 또는 위쪽/아래쪽을 사용할 때) 적용된다. 값에는 포인트 단위가 있다.
